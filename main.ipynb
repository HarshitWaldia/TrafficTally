## Vehicle Detection And Counting

```python
import cv2

import numpy as np 
```

## Web Cam

```python
cap=cv2.VideoCapture("video3.mp4")
```

## Count Line Position / MIN_WIDTH / MIN_HEIGHT

```python

count_line_position=550 

min_width_rectangle=80

min_height_rectangle=80
```

## Algorithm : BGSEGM Substractor i.e. Background Substractor

```python
algo=cv2.bgsegm.createBackgroundSubtractorMOG()
```

## Center Function

```python
def center_handle(x,y,w,h):

    x1=int(w/2)

    y1=int(h/2)

    cx=x+x1

    cy=y+y1

    return cx,cy
```

## Empty List

```python
detect=[]

offset=6

count=0
```

```python
while True:

    ret,frame=cap.read()
    
    BG_Grey=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    blur=cv2.GaussianBlur(BG_Grey,(3,3),5)

    # applying on each frame
    
    img_sub=algo.apply(blur)
    
    dilat=cv2.dilate(img_sub,np.ones((5,5)))

    kernal=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
    
    dilite_ada=cv2.morphologyEx(dilat,cv2.MORPH_CLOSE,kernal)
    
    dilite_ada=cv2.morphologyEx(dilite_ada,cv2.MORPH_CLOSE,kernal)
    
    contour_shape,h = cv2.findContours(dilite_ada,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    
    cv2.line(frame,(25,count_line_position),(1200,count_line_position),(255,127,0),3)
    

    for (i,c) in enumerate(contour_shape):
    
        (x,y,w,h)=cv2.boundingRect(c)
    
        validation_counter= (w>=min_width_rectangle) and (h>=min_height_rectangle)
    
        if not validation_counter:
    
            continue
    
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,225,0),2)
    
        cv2.putText(frame,"VEHICLE_COUNTER = "+str(count),(x,y-20),cv2.FONT_HERSHEY_SIMPLEX,1,(255,244,0),2)


        center=center_handle(x,y,w,h)
    
        detect.append(center)
    
        cv2.circle(frame,center,4,(0,0,255),-2)

        for (x,y) in detect:
    
            if y<(count_line_position+offset) and y>(count_line_position-offset):
    
                count+=1
    
                cv2.line(frame,(25,count_line_position),(1200,count_line_position),(0,127,255),3)
    
                detect.remove((x,y))
    
            print("VEHICLE_COUNT = "+str(count))
    
    cv2.putText(frame,"VEHICLE_COUNTER = "+str(count),(450,70),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),5)

    #cv2.imshow("detector",dilite_ada) #dilation 
    
    cv2.imshow("Detection video",frame) #main frame

    if cv2.waitKey(1) == ord("q"):
    
        break

cv2.destroyAllWindows()

cap.release()
```
